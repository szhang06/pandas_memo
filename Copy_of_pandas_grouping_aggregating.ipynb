# Group: split and combine

```python
df["Country"].value_counts(normalize=True)
```

```python
country_grp = df.groupby(["Country"])
```

```python
country_grp.get_group("United States")
```

```python
filt = df["Country"] =="China"
df.loc[filt]["SocialMedia"].value_counts()
```

```python
country_grp["SocialMedia"].value_counts().loc["India"]
```

```python
country_grp["ConvertedComp"].median().loc["Germany"]
```

```python
country_grp["ConvertedComp"].agg(["median", "mean"]).loc["Canada"]
```

```python
filt= (df["Country"] == "India")
df.loc[filt]["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
#country_grp["LanguageWorkedWith"].str.contains("Python").sum() #not worked
```

```python
country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
```

```python
country_respondents = df["Country"].value_counts()
country_respondents
```

```python
country_uses_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum()) # apply a function along an axis of the DataFrame
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis = "columns", sort=False)
python_df
```

```python
# python_df_row = pd.concat([country_respondents, country_uses_python], sort=False) # default axis="rows", concat along rows
# python_df_row
```

```python
python_df.rename(columns={"Country": "NumRespondents", "LanguageWorkedWith": "NumKnowsPython"}, inplace = True)
python_df
```

```python
python_df["PetKnowsPython"] = (python_df["NumKnowsPython"]/python_df["NumRespondents"])*100
```

```python
python_df
```

```python
python_df.sort_values(by="PetKnowsPython", ascending=False, inplace=True)
python_df.loc["China"]
```

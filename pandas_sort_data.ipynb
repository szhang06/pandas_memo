```python
people = {"first":["Corey", "Jane", "John", "Adam"],
          "last": ["Schafer", "Doe", "Doe", "Doe"],
          "email":["CoreyMSchafter@gamil.com", "JaneDoe@email.com", "JohnDoe@email.com", "Ad@email.com"]}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
df.sort_values(by="last", ascending=False) 
```

```python
df.sort_values(by=["last", "first"], ascending=[False, True], inplace = True)  # sort by "last" descending and then "first" by ascending; 
```

```python
df # index order changes accordingly
```

```python
df.sort_index() # return to the original, index sorting
```

```python
# sort single column
df["last"].sort_values()
```
